```python
from collections import defaultdict
import random

def set_num():
    print("숫자를 설정합니다.")
    com_num = str(random.randint(100,999))
    com_dict = {i:int(com_num[i]) for i in range(len(com_num))}
    
    return com_dict

def input_num():
    user_num = input("숫자를 입력하세요 : ")
    user_dict = {i : int(user_num[i]) for i in range(len(user_num))}
    
    return user_dict

def compare(set, input):
    ball = 0
    strike = 0
    
    concat = ''
    
    for i in set.values():
        concat += str(i)
    
    for i in enumerate(set):
        for j in enumerate(input):
            if i[0] == j[0] and set[i[1]] == input[j[1]]:
                strike += 1
                break;
                
    for i in input.values():
        if concat.find(str(i)) != -1:
            ball +=1
            
    return [strike, ball]
    
def verdict(result, com):
    if result[0] == 3 :
        concat = ''
        for i in com.values():
            concat+= str(i)
        print("정답입니다! 정답은 " + concat + "입니다.")
        return 0
    else :
        print("틀렸습니다! " + str(result[0]) + "스트라이트, " + str(result[1]) + "볼 입니다.")
        return 1
    
    
class Game:
            
    com = 0
    user = 0
    cnt = 1
        
    def start(cnt):
        com = set_num()
    
        while(cnt):
            user = input_num()
            cnt = verdict(compare(com, user), com)
```
